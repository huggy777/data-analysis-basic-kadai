```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2,3,tight_layout = True)

#折れ線グラフ
y0 = [1,2,-5,2]
axes[0,0].plot(y0)

#sin関数
x1 = np.arange(0,6,0.1)
y1 = np.sin(2*x1)
axes[0,1].plot(x1,y1)

#ヒストグラム
y2 = np.random.normal(10,2,1000)
axes[1,0].hist(y2,bins=20)

#散布図
x3 = np.random.randint(10,100,10)
y3 = np.random.randint(0.1,10,10)
axes[1,1].scatter(x3,y3)

plt.show()
```
